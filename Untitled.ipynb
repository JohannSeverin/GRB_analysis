```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.signal import medfilt
%matplotlib notebook
```

```python
def plot_lc(grbname,axes,filt=1):
    names = ['Time','b1','b1er','b2','b2er','b3','b3er','b4','b4er','b5','b5er']
    lc = pd.read_csv("BAT_lc/%s_lc.dat"%grbname,sep = ' ',names=names)
    for band, colour in zip(['b1','b2','b3','b4'],['k','r','g','b']):
        axes.plot(lc.Time,medfilt(lc.loc[:,band],filt),c=colour,alpha=1)
        axes.set_title(grbname)
    return
```

```python
fig, ax = plt.subplots(3,sharex=True)
grbs = ['GRB091018','GRB111121A','GRB060313']
for i in range(len(ax)):
    plot_lc(grbs[i],ax[i])
```

```python
ax[0].set_xlim(-4,10)
```

```python
for i in range(3):
    ax[i].tick_params(axis='y',right=True,left=False,labelright=True,labelleft=False)
ax[1].set_ylabel('counts/sec/det',rotation=-90,labelpad=-385)
ax[2].set_xlabel('Time since BAT-trigger $[s]$')
```
